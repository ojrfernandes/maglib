```python
## Plot connection length
# Import required modules
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# Load data
data = np.loadtxt('test_reading_00.dat')

# Extract relevant columns
x = data[:, 2] #phi
y = data[:, 0] #r_0
z = data[:, 7] #color


# Determine the grid size
num_rows = int(len(np.unique(x))) # Number of rows determined by the number of unrepeated phi values
num_columns = int(len(data)/num_rows) # Number of columns determined by the data's length and the number of rows
z = np.reshape(z, (num_rows, num_columns)).transpose() # Convert z column to matrix
z = np.tile(z, (1, 3)) # Replicate the first 120 degrees three times along the x-axis to complete 360 (for perfect n=3)


# Image plot
plt.figure(figsize=(9, 6))
plt.imshow(z, cmap='jet', extent=[0, 360, np.min(y), np.max(y)], origin='lower', aspect='auto', norm=LogNorm())
plt.colorbar().set_label('connection length [m]')

# Set intervals for the grid
x_min, x_max = 0, 360
y_min, y_max = 0.471, 0.477

num_x_divisions = 14
num_y_divisions = 10

# Generate x and y values for the grid
x_grid = np.linspace(x_min, x_max, num_x_divisions)
y_grid = np.linspace(y_min, y_max, num_y_divisions + 1)


# Plot the grid
#plt.grid(True, linestyle='--', linewidth=0.5, color='black', alpha=0.7)

# Set x and y axis ranges
#plt.xlim(x_min, x_max)
#plt.ylim(y_min, y_max)

# Set ticks for the grid
#plt.xticks(x_grid)
#plt.yticks(y_grid)


# Labels
plt.xlabel('$\phi$ [m]')
plt.ylabel('$R$ [m]')
plt.title('Magnetic Footprints')

# Show plot
plt.show()
```
